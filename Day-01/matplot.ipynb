```python
import matplotlib.pyplot as plt
import numpy as np

# Set style
plt.style.use('default')

# Create a figure with multiple subplots
fig = plt.figure(figsize=(16, 12))

# 1. Line Chart
ax1 = plt.subplot(3, 3, 1)
x = np.linspace(0, 10, 100)
ax1.plot(x, np.sin(x), label='sin(x)')
ax1.plot(x, np.cos(x), label='cos(x)')
ax1.set_title('Line Chart')
ax1.legend()
ax1.grid(True)

# 2. Scatter Plot
ax2 = plt.subplot(3, 3, 2)
x = np.random.rand(50)
y = np.random.rand(50)
ax2.scatter(x, y, c=np.random.rand(50), cmap='viridis', s=100, alpha=0.6)
ax2.set_title('Scatter Plot')

# 3. Bar Chart
ax3 = plt.subplot(3, 3, 3)
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.rand(5) * 10
ax3.bar(categories, values, color='skyblue', edgecolor='navy')
ax3.set_title('Bar Chart')

# 4. Histogram
ax4 = plt.subplot(3, 3, 4)
data = np.random.normal(100, 15, 1000)
ax4.hist(data, bins=30, color='lightcoral', edgecolor='black')
ax4.set_title('Histogram')

# 5. Pie Chart
ax5 = plt.subplot(3, 3, 5)
sizes = [30, 25, 20, 15, 10]
labels = ['A', 'B', 'C', 'D', 'E']
ax5.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
ax5.set_title('Pie Chart')

# 6. Box Plot
ax6 = plt.subplot(3, 3, 6)
data = [np.random.normal(0, std, 100) for std in range(1, 5)]
ax6.boxplot(data, labels=['A', 'B', 'C', 'D'])
ax6.set_title('Box Plot')

# 7. Area Chart
ax7 = plt.subplot(3, 3, 7)
x = np.linspace(0, 10, 50)
y1 = np.sin(x)
y2 = np.cos(x)
ax7.fill_between(x, y1, alpha=0.5, label='sin(x)')
ax7.fill_between(x, y2, alpha=0.5, label='cos(x)')
ax7.set_title('Area Chart')
ax7.legend()

# 8. Heatmap
ax8 = plt.subplot(3, 3, 8)
data = np.random.rand(10, 10)
im = ax8.imshow(data, cmap='hot', aspect='auto')
ax8.set_title('Heatmap')
plt.colorbar(im, ax=ax8)

# 9. Contour Plot
ax9 = plt.subplot(3, 3, 9)
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))
contour = ax9.contour(X, Y, Z, levels=15, cmap='viridis')
ax9.set_title('Contour Plot')

plt.tight_layout()
plt.show()
```
