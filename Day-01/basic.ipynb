```python
vars_to_show = [
    "age", "coordinates", "height", "is_student", "name",
    "numbers", "person", "unique_numbers", "value"
]

for var in vars_to_show:
    val = globals().get(var)
    tname = type(val).__name__
    try:
        length = len(val)
    except TypeError:
        print(f"{var}: type={tname}, value={val}")
    else:
        print(f"{var}: type={tname}, value={val}, length={length}")
```

```python
# control flow examples using existing notebook variables

if age >= 18:
    print(f"{name} is an adult (age={age}).")
else:
    print(f"{name} is a minor (age={age}).")

if is_student:
    print(f"{name} is a student.")

if height > 6:
    print("Tall")
else:
    print("Not very tall")

if numbers:
    if len(numbers) == length:
        print(f"numbers length matches expected {length}.")
    else:
        print(f"numbers length ({len(numbers)}) does NOT match expected {length}.")

if 'name' in person and person.get('name') == 'Bob':
    print("Person is Bob")

if value is None:
    print("value is None")

if 3 in unique_numbers:
    print("3 is in unique_numbers")
```

```python
# if-elif-else examples using existing notebook variables

# age classification
if age < 13:
    print(f"{name} is a child (age={age}).")
elif age < 20:
    print(f"{name} is a teenager (age={age}).")
elif age < 65:
    print(f"{name} is an adult (age={age}).")
else:
    print(f"{name} is a senior (age={age}).")

# height category
if height >= 7:
    print("Very tall")
elif height >= 6:
    print("Tall")
elif height >= 5:
    print("Average height")
else:
    print("Short")

# student status and numbers length check
if is_student and len(numbers) == length:
    print(f"{name} is a student and numbers length matches expected {length}.")
elif is_student:
    print(f"{name} is a student but numbers length ({len(numbers)}) does not match expected {length}.")
elif len(numbers) == length:
    print(f"{name} is not a student but numbers length matches expected {length}.")
else:
    print(f"{name} is not a student and numbers length ({len(numbers)}) does not match expected {length}.")

# check person name
if 'name' in person and person.get('name') == name:
    print("Person name matches notebook name")
elif 'name' in person and person.get('name') == 'Bob':
    print("Person is Bob")
else:
    print("Person is someone else")
```
