```python
import re

# Sample text
text = "Hello! This is a sample text. We will tokenize it into words and sentences."

# Tokenize into sentences using regex
sentences = re.split(r'[.!?]+', text)
sentences = [s.strip() for s in sentences if s.strip()]
print("Sentence tokens:")
print(sentences)

# Tokenize into words using regex
words = re.findall(r'\b\w+\b', text.lower())
print("\nWord tokens:")
print(words)

# Tokenize with punctuation preserved
words_with_punct = re.findall(r'\b\w+\b|[.!?,;:]', text.lower())
print("\nWord tokens (with punctuation):")
print(words_with_punct)
```
