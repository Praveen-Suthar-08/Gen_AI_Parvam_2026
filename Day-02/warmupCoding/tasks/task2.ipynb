```python
import re

def jaccard_similarity(sentence1: str, sentence2: str) -> float:
    def tokenize(text: str) -> set[str]:
        words = re.findall(r"\b\w+\b", text.lower())
        return set(words)

    set1 = tokenize(sentence1)
    set2 = tokenize(sentence2)

    if not set1 and not set2:
        return 1.0
    if not set1 or not set2:
        return 0.0
    intersection = set1 & set2
    
    union = set1 | set2
    return len(intersection) / len(union)

def are_sentences_similar(sentence1: str, sentence2: str, threshold: float = 0.5) -> bool:
    return jaccard_similarity(sentence1, sentence2) >= threshold

# Test cases
print("=== Jaccard Similarity Tests ===\n")

# Test 1: Similar sentences
s1 = "The quick brown fox"
s2 = "The quick brown dog"
sim = jaccard_similarity(s1, s2)
print(f"Sentence 1: '{s1}'")
print(f"Sentence 2: '{s2}'")
print(f"Similarity: {sim:.2f}")
print(f"Similar (threshold=0.5): {are_sentences_similar(s1, s2, 0.5)}\n")

# Test 2: Identical sentences
s3 = "hello world"
s4 = "hello world"
sim2 = jaccard_similarity(s3, s4)
print(f"Sentence 1: '{s3}'")
print(f"Sentence 2: '{s4}'")
print(f"Similarity: {sim2:.2f}")
print(f"Similar (threshold=0.5): {are_sentences_similar(s3, s4, 0.5)}\n")

# Test 3: Completely different
s5 = "abc def"
s6 = "xyz uvw"
sim3 = jaccard_similarity(s5, s6)
print(f"Sentence 1: '{s5}'")
print(f"Sentence 2: '{s6}'")
print(f"Similarity: {sim3:.2f}")
print(f"Similar (threshold=0.5): {are_sentences_similar(s5, s6, 0.5)}\n")

# Test 4: Empty strings
sim4 = jaccard_similarity("", "")
print(f"Empty strings similarity: {sim4:.2f}")
```
