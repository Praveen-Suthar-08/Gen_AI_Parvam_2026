```python
def classify_sentence(sentence):
    """
    Classifies a sentence as question, exclamation, or statement.
    
    Args:
        sentence (str): The sentence to classify
        
    Returns:
        str: The classification ('question', 'exclamation', or 'statement')
    """
    sentence = sentence.strip()
    
    if sentence.endswith('?'):
        return 'question'
    elif sentence.endswith('!'):
        return 'exclamation'
    else:
        return 'statement'


# Test examples
test_sentences = [
    "What is your name?",
    "I love programming!",
    "This is a statement",
    "How are you?",
    "Amazing!"
]

for sentence in test_sentences:
    classification = classify_sentence(sentence)
    print(f"'{sentence}' -> {classification}")
```
