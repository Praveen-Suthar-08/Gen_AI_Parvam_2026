```python
import string

def tokenize_sentence(sentence):
    """
    Tokenizes a sentence into words and removes punctuation.
    
    Args:
        sentence (str): The input sentence to tokenize
        
    Returns:
        list: A list of words without punctuation
    """
    # Remove punctuation and convert to lowercase
    translator = str.maketrans('', '', string.punctuation)
    cleaned = sentence.translate(translator).lower()
    
    # Split into words
    words = cleaned.split()
    
    return words
```
