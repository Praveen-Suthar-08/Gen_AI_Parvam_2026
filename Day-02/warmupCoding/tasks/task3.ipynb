```python
from collections import Counter
import re
from nltk.corpus import stopwords
import nltk

# Download stopwords if not already present
nltk.download('stopwords', quiet=True)

def extract_keywords(text, n=10):
    """
    Extract top N most frequent meaningful words from text.
    
    Args:
        text (str): Input text to extract keywords from
        n (int): Number of top keywords to return
    
    Returns:
        list: Top N most frequent meaningful words
    """
    # Convert to lowercase and remove special characters
    text = text.lower()
    words = re.findall(r'\b[a-z]+\b', text)
    
    # Get English stopwords
    stop_words = set(stopwords.words('english'))
    
    # Filter out stopwords
    meaningful_words = [word for word in words if word not in stop_words]
    
    # Count frequencies and get top N
    word_counts = Counter(meaningful_words)
    top_keywords = word_counts.most_common(n)
    
    return [word for word, count in top_keywords]

# Example usage
sample_text = "Python is great for data science. Python is powerful and Python is easy to learn."
keywords = extract_keywords(sample_text, n=5)
print(keywords)
```
