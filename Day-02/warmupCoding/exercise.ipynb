```python
import re
from collections import Counter

class TextAnalyzer:
    def __init__(self, text):
        self.text = text
    
    def word_count(self):
        """Count total words in text"""
        words = self.text.lower().split()
        return len(words)
    
    def unique_words(self):
        """Count unique words"""
        words = re.findall(r'\b\w+\b', self.text.lower())
        return len(set(words))
    
    def character_count(self):
        """Count total characters"""
        return len(self.text)
    
    def sentence_count(self):
        """Count sentences"""
        sentences = re.split(r'[.!?]+', self.text)
        return len([s for s in sentences if s.strip()])
    
    def word_frequency(self, top_n=10):
        """Get most common words"""
        words = re.findall(r'\b\w+\b', self.text.lower())
        return Counter(words).most_common(top_n)
    
    def average_word_length(self):
        """Calculate average word length"""
        words = re.findall(r'\b\w+\b', self.text.lower())
        if not words:
            return 0
        return sum(len(word) for word in words) / len(words)
    
    def summary(self):
        """Print analysis summary"""
        print(f"Words: {self.word_count()}")
        print(f"Unique Words: {self.unique_words()}")
        print(f"Characters: {self.character_count()}")
        print(f"Sentences: {self.sentence_count()}")
        print(f"Average Word Length: {self.average_word_length():.2f}")

# Example usage
sample_text = "Python is great. Python is fun. I love Python programming!"
analyzer = TextAnalyzer(sample_text)
analyzer.summary()
print("\nTop 5 Words:", analyzer.word_frequency(5))
```
